```python
dt = dt.drop(columns = "Unnamed")
```

```python
dt["y"] = dt["y"].astype("int")
dt["y"] = dt["y"].apply(lambda x : 0 if x == 1 else 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(dt.drop(columns=["y"]),dt["y"],test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.ensemble import AdaBoostClassifier,  RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
RandomForest = make_pipeline(StandardScaler(), RandomForestClassifier(random_state=0))
AdaBoost = make_pipeline(StandardScaler(), AdaBoostClassifier(random_state=0))
KNN = make_pipeline(StandardScaler(),  KNeighborsClassifier())
```

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
```

```python
def evaluation(model):
    
    model.fit(x_train, y_train)
    ypred = model.predict(x_test)
    
    sns.heatmap(confusion_matrix(y_test, ypred),annot=True)
    print(classification_report(y_test, ypred))
    
    N, train_score, val_score = learning_curve(model, x_train, y_train,
                                              cv=3,
                                               train_sizes=np.linspace(0.1, 1, 5))
    
    
    plt.figure(figsize=(12, 8))
    plt.plot(N, train_score.mean(axis=1), label='train score')
    plt.plot(N, val_score.mean(axis=1), label='validation score')
    plt.legend()
```

```python
evaluation(RandomForest)
```

```python
evaluation(AdaBoost)
```

```python
evaluation(KNN)
```

```python
x_train = StandardScaler().fit_transform(x_train)
x_test = StandardScaler().fit_transform(x_test)
```

```python
x_train.shape,y_train.shape
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(160,activation="relu",kernel_regularizer=tf.keras.regularizers.l2(0.001), input_dim = 178))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(160,kernel_regularizer=tf.keras.regularizers.l2(0.001), activation = "relu"))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(80,kernel_regularizer=tf.keras.regularizers.l2(0.001),activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(40, kernel_regularizer=tf.keras.regularizers.l2(0.001),activation="relu"))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(1, activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss=tf.keras.losses.binary_crossentropy, 
                optimizer=tf.keras.optimizers.RMSprop(), 
                metrics=["accuracy"])
```

```python
history = model.fit(x_train,y_train,
                      validation_split=0.2,
                    batch_size =10,
                      epochs=100)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training and Validation accuracy')
plt.legend(['Training accuracy', 'Validation accuracy'])
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print("accuracy : ", accuracy)
```
